```python
def read_epub_sentences(epub_file, epub_ID):
    book = epub.read_epub(epub_file)
    sentences = []
    
    # Extract content from each document item in the EPUB file
    for item in book.get_items_of_type(ebooklib.ITEM_DOCUMENT):
        content = item.get_content().decode('utf-8')
        content = re.sub('<[^<]+?>', '', content)  # Remove HTML tags
        content = re.sub('\s+', ' ', content)     # Replace multiple spaces with a single space
        content = re.sub('\n', ' ', content)      # Replace newlines with space
        content = re.sub('&#13;', ' ', content)  # Replace carriage returns with space
        
        # Split content into sentences by period (".")
        sentences.extend(content.strip().split("."))
    
    # Create a list of sentences with their respective number and book ID
    sentences = [{'Paragraph': sentences[i].strip(), 'Number': i, 'Book ID': epub_ID} for i in range(len(sentences))]
    
    # Return sentences, excluding the first and last empty ones if any
    return sentences[1:-1]
```

```python
def merge_strings_until_limit(strings, min_length, max_length, test_for_max=0):
    merged_string = ""
    merged_strings = []
    
    for s in strings:
        # If the merged string is still within the min_length limit, continue adding sentences
        if len(merged_string) <= min_length:
            merged_string += s
        
        # If merged string exceeds the max_length limit, split and recursively merge
        elif len(merged_string) > max_length and test_for_max < 5:
            splitParagraph = merged_string.split('.')
            splitParagraphRePoint = [sp + '.' for sp in splitParagraph]  # Add period back
            merged = merge_strings_until_limit(splitParagraphRePoint, min_length, max_length, test_for_max + 1)
            merged_strings.extend(merged)
            merged_string = s
        else:
            # Otherwise, store the current merged string and start a new one with the current sentence
            merged_strings.append(merged_string)
            merged_string = s
    
    if merged_string:
        merged_strings.append(merged_string)
    
    return merged_strings
```

```python
def read_epub_paragraphs(epub_file, epub_ID):
    book = epub.read_epub(epub_file)
    paragraphs = []
    
    # Extract content from each document item in the EPUB file
    for item in book.get_items_of_type(ebooklib.ITEM_DOCUMENT):
        content = item.get_content().decode('utf-8')
        content = re.sub('<[^<]+?>', '', content)  # Remove HTML tags
        content = re.sub('\s+', ' ', content)     # Replace multiple spaces with a single space
        content = re.sub('\n', ' ', content)      # Replace newlines with space
        content = re.sub('&#13;', ' ', content)  # Replace carriage returns with space
        
        # Split content into paragraphs by the special string (e.g., &#13;)
        paragraphs.extend(content.strip().split("&#13;"))
    
    # Merge the paragraphs within the specified length limit
    paragraphs = merge_strings_until_limit(paragraphs, 200, 1000)
    
    # Create a list of paragraphs with their respective number and book ID
    paragraphs = [{'Paragraph': paragraphs[i], 'Number': i, 'Book ID': epub_ID} for i in range(len(paragraphs))]
    
    # Return paragraphs, excluding the first and last empty ones if any
    return paragraphs[1:-1]
```

```python
# Loop through each EPUB file in the directory
epub_files = [f for f in os.listdir(epub_dir) if f.endswith('.epub')]

# Loop through and display content for each EPUB file
for epub_file in epub_files:
    epub_file_path = os.path.join(epub_dir, epub_file)
    print(f"\nReading: {epub_file_path}")
    
    # Extract paragraphs from each EPUB file
    paragraphs = read_epub_paragraphs(epub_file_path, epub_file)
    
    # Display paragraphs 10 to 16 (adjust the range as needed)
    for i in range(9, 16):  # Display paragraphs 10 to 16 (index 9 to 15)
        if i < len(paragraphs):
            paragraph = paragraphs[i]
            print(f"Paragraph {paragraph['Number']} from Book {paragraph['Book ID']}:")
            print(paragraph['Paragraph'])
            print("\n" + "-"*50 + "\n")
        else:
            print(f"Paragraph {i + 1} does not exist in Book {epub_file}.\n")

print("\nContent extraction and merging complete.")
```
