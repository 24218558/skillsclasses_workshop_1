```python
# Compare search methods
def compare_methods(query, num_results=5):
    methods = ['simple', 'tfidf', 'svd']
    results = {method: search_database(query, method, num_results) for method in methods}
    
    for method, result_set in results.items():
        print(f"Results for {method} method:")
        
        if not isinstance(result_set, list):
            print(f"Error: Expected a list of results, but got {type(result_set)} for method '{method}'")
            continue
        
        for result in result_set:
            # Check if the result is a dictionary with the expected keys
            if isinstance(result, dict) and {'BOOK', 'LINE', 'TEXT', 'KEYWORD'}.issubset(result):
                print(f"BOOK: {result['BOOK']}, LINE: {result['LINE']}\nTEXT: {result['TEXT']}\nKEYWORD: {result['KEYWORD']}\n")
            else:
                print(f"Warning: Unexpected result format in method '{method}': {result}")
        
        print("\n" + "-" * 40 + "\n")
```

```python
# Example Usage
query = input("Enter your search query: ")
num_results = int(input("Enter number of results: "))
compare_methods(query, num_results)
```
